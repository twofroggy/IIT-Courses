```python
#1.1.12a 
#declare variables
n = 0  #monthly payments for 2 years = 24 months 
balance = 12000
p = 0

#while loop for when credit card balance is greater than 0
while balance > 0: 
    #reset balance to 12000 so it's not infinitely growing
    balance = 12000 
    #for loop 24 months
    for n in range(24):
        balance = (1 + (0.199/12)) * balance - p
    #increment p by 1 cent to get most accurate payment amount
    p = p + 0.01  
        
#print the monthly payment, p, value that makes balance = 0
print (p) 
```

```python
#1.2.2 
#import pyplot
from matplotlib import pyplot as plt
import numpy as np 

#create arrays for each variable
year = []
population = [3929000, 5308000, 7240000, 9638000, 12866000, 17069000, 23192000, 
              31443000, 38558000, 50156000, 62948000, 75995000, 91972000, 
              105711000,122755000, 131669000, 150697000, 179323000, 203212000, 
              226505000, 248710000, 281416000, 308746000]

#fill year array with for loop
for i in range(23):
    year.append(i)

#create a scatter plot    
plt.scatter(year, population)
plt.xlabel('Year since 1970 (n, in 10s)')
plt.ylabel('Population (P(n), in millions)')
plt.show()
```

```python
#import pyplot
from matplotlib import pyplot as plt
import numpy as np 

#create arrays for each variable
change = [] #delta(population array)
    
#fill an array with delta(population)
for i in range(1, len(population)):
    past = population[i-1] 
    future = population[i] 
    change.append(future - past)

#new array for p(n+1) - p(n) 
pop = [5308000, 7240000, 9638000, 12866000, 17069000, 23192000, 
       31443000, 38558000, 50156000, 62948000, 75995000, 91972000, 
       105711000,122755000, 131669000, 150697000, 179323000, 203212000, 
       226505000, 248710000, 281416000, 308746000]

#create a scatter plot    
plt.scatter(pop, change)
plt.xlabel('Population P(n)')
plt.ylabel('Delta(P(n)) (in millions)')
plt.show()
```

```python
#create a line of best fit of the model delta(p(n)) vs p(n)
x = np.array(pop)
y = np.array(change)
m, b = np.polyfit(pop, change, 1)
plt.plot(pop, change, 'o')
plt.plot(x, m*x + b)
print("slope = ", m, " and y-intercept = ", b)
```

```python
#plot actual data and modeled data on the same graph to compare
ax = plt.gca()

ax.scatter(year, population, color="b", label="actual data")
ax.scatter(year, predictedpop, color="r", label="modeled data")
plt.title("Population vs Year (P(n) vs n)")
plt.xlabel("Year (n)")
plt.ylabel("Population (P(n))")
plt.legend(loc="upper left")
plt.show()
```
