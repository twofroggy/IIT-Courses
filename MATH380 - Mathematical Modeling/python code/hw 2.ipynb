# 1.3.3a


```python
#given: a(n+1) = -1.2(a(n)) + 50, and a(0) = 1000
n = []
a = []
greater = []
lesser = []
equal = []

start = 1000
for i in range(25):
    a.append(-1.2*start + 50)
    n.append(i)
    start = a[i]
    #print('actual', a[i])

start = 22.727272
for i in range(25):
    equal.append(-1.2*start + 50)
    start = equal[i] 
    #print('equal', equal[i])

start = 1500
for i in range(25):
    greater.append(-1.2*start + 50)
    start = greater[i] 
    #print('greater', greater[i])
    
start = 500
for i in range(25):
    lesser.append(-1.2*start + 50) 
    start = lesser[i]
    #print('lesser', lesser[i])
    
import pandas as pd

data = {'n': n, 'actual data': a, 
        'equillibrium data': equal, 
        'greater starting value': greater, 
        'lesser starting value': lesser}
df = pd.DataFrame(data=data)
df
```

```python
#actual data
plt.scatter(n, a)
plt.xlabel('n')
plt.ylabel('a(n)')
plt.show()
```

```python
ax = plt.gca()

ax.scatter(n, greater, color="r", label="a(0) > equillibrium value")
ax.scatter(n, lesser, color="g", label="a(0) < equillibrium value")
ax.scatter(n, equal, color="y", label="a(0) = equillibrium value")
plt.title("n vs a(n)")
plt.xlabel("n")
plt.ylabel("a(n)")
plt.legend(loc="upper left")
plt.show() 
```
